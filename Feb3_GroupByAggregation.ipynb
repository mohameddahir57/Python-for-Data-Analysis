In this session, we will analyze a sales dataset using GroupBy operations.

```python
import pandas as pd

# Sample sales data
data = {
    'Region': ['East', 'West', 'East', 'South', 'West', 'North', 'South', 'North'],
    'Product': ['Apple', 'Orange', 'Apple', 'Banana', 'Orange', 'Apple', 'Banana', 'Orange'],
    'Sales': [1000, 1500, 1200, 800, 1400, 2000, 900, 1100],
    'Profit': [100, 200, 150, 50, 180, 300, 60, 110]
}

df = pd.DataFrame(data)
print("Sales Data:")
df
```

### 1. Basic GroupBy
Find total sales per Region.

```python
region_sales = df.groupby('Region')['Sales'].sum()
print("Total Sales per Region:")
print(region_sales)
```

### 2. Multi-column GroupBy
Find average sales per Region and Product.

```python
multi_group = df.groupby(['Region', 'Product'])['Sales'].mean()
print("Average sales per Region and Product:")
print(multi_group)
```

### 3. Using `.agg()` for multiple metrics

```python
summary = df.groupby('Product').agg({
    'Sales': ['sum', 'mean'],
    'Profit': 'sum'
})
print("Product Summary Statistics:")
summary
```

### 4. Simple Pivot Table

```python
pivot = df.pivot_table(index='Region', columns='Product', values='Profit', aggfunc='sum', fill_value=0)
print("Profit Pivot Table (Region vs Product):")
pivot
```
