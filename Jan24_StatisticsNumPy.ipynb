Calculate basic statistics like mean, median, standard deviation, and more.

```python
# Import the NumPy library and alias it as 'np'
import numpy as np
```

## 1. Basic Statistical Functions

```python
# Create a dataset representing test scores
scores = np.array([85, 90, 78, 92, 88, 76, 95, 89])
print(f"Scores: {scores}")

# Calculate the mean (average) score using np.mean()
mean_score = np.mean(scores)
print(f"Mean: {mean_score}")

# Calculate the median (middle) score using np.median()
median_score = np.median(scores)
print(f"Median: {median_score}")

# Calculate the standard deviation (how spread out the scores are) using np.std()
std_dev = np.std(scores)
print(f"Standard Deviation: {std_dev:.2f}") # .2f formats to 2 decimal places

# Find the minimum score using np.min()
min_score = np.min(scores)
print(f"Min Score: {min_score}")

# Find the maximum score using np.max()
max_score = np.max(scores)
print(f"Max Score: {max_score}")
```

## 2. Statistics with Axes (2D Arrays)

```python
# Create a 2D array representing scores for 3 students across 3 subjects
# Rows = Students, Columns = Subjects
class_scores = np.array([
    [85, 90, 88], # Student 1
    [78, 82, 80], # Student 2
    [92, 95, 94]  # Student 3
])
print(f"Class Scores:\n{class_scores}")

# Calculate the average score for each subject (average of each column, axis=0)
subject_means = np.mean(class_scores, axis=0)
print(f"Average per Subject (axis=0): {subject_means}")

# Calculate the average score for each student (average of each row, axis=1)
student_means = np.mean(class_scores, axis=1)
print(f"Average per Student (axis=1): {student_means}")
```
