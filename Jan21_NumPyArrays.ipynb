Practice using NumPy to create and inspect arrays.

```python
# Import the NumPy library and alias it as 'np' for convenience
import numpy as np
```

## 1. Creating Arrays from Lists

```python
# Print a header for the section
print("--- 1. Creating Arrays from Lists ---")

# Create a standard Python list containing integers from 1 to 5
my_list = [1, 2, 3, 4, 5]

# Convert the Python list into a NumPy array using np.array()
arr = np.array(my_list)

# Print the original Python list
print(f"List: {my_list}")

# Print the newly created NumPy array to see the difference in output format
print(f"NumPy Array:\n{arr}")
```

## 2. Creating Special Arrays

```python
# Print a header for the section
print("--- 2. Creating Special Arrays ---")

# Create a 2x3 array (2 rows, 3 columns) filled with zeros
zeros = np.zeros((2, 3)) 

# Create a 3x2 array (3 rows, 2 columns) filled with ones
ones = np.ones((3, 2))   

# Create an array starting at 0, stopping before 20, with a step size of 2 (0, 2, 4, ... 18)
range_arr = np.arange(0, 20, 2) 

# Create an array with 5 evenly spaced values between 0 and 10 (inclusive)
linspace_arr = np.linspace(0, 10, 5) 

# Print the array of zeros
print(f"Zeros Array:\n{zeros}")

# Print the array of ones
print(f"Ones Array:\n{ones}")

# Print the array created with arange
print(f"Arange Array: {range_arr}")

# Print the array created with linspace
print(f"Linspace Array: {linspace_arr}")
```

## 3. Array Attributes

```python
# Print a header for the section
print("--- 3. Array Attributes ---")

# Create a 2D array (matrix) from a list of lists
matrix = np.array([[1, 2, 3], [4, 5, 6]])

# Print the matrix content
print(f"Matrix:\n{matrix}")

# Print the number of dimensions (axes) of the array. For a 2D matrix, this is 2.
print(f"Dimensions (.ndim): {matrix.ndim}")

# Print the shape of the array as a tuple (rows, columns)
print(f"Shape (.shape): {matrix.shape}")

# Print the total number of elements in the array (rows * columns)
print(f"Total Elements (.size): {matrix.size}")

# Print the data type of the elements inside the array (e.g., int32 or int64)
print(f"Data Type (.dtype): {matrix.dtype}")
```
