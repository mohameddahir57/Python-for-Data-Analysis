In this session, we will clean a dataset where types are incorrectly assigned.

```python
import pandas as pd

# Sample dirty dataset
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Price': ['10.5', '20', '30.25', 'N/A'],
    'Quantity': ['1', '5', '3', '2'],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04']
}

df = pd.DataFrame(data)
print("Initial Data Types:")
print(df.dtypes)
df
```

### 1. Fix Numeric Types
Use `pd.to_numeric` with `errors='coerce'` for the Price column because of the 'N/A' value.

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Quantity'] = df['Quantity'].astype(int)

print("Types after numeric conversion:")
print(df.dtypes)
df
```

### 2. Fix Date Types

```python
df['Date'] = pd.to_datetime(df['Date'])
print("Final Types:")
print(df.dtypes)
df
```

### 3. Extracting info from Dates

```python
df['Day'] = df['Date'].dt.day_name()
df
```
