Let's practice finding and removing duplicates in a dataset.

```python
import pandas as pd

# Sample data with duplicates
data = {
    'ID': [101, 102, 101, 103, 102, 104],
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'David'],
    'Score': [85, 90, 85, 95, 88, 70]
}

df = pd.DataFrame(data)
print("Original Data:")
df
```

### 1. Count duplicate rows

```python
print(f"Total duplicates: {df.duplicated().sum()}")
print("Duplicate rows:")
df[df.duplicated()]
```

### 2. Remove exact duplicates

```python
df_clean = df.drop_duplicates()
print("Data after removing exact duplicates:")
df_clean
```

### 3. Remove duplicates based on a specific column
Notice that ID 102 has different scores (90 and 88). If we only care about the unique ID, we must decide which one to keep.

```python
# Keep the last entry for each ID
df_unique_id = df.drop_duplicates(subset=['ID'], keep='last')
print("Data keeping last occurrence of each ID:")
df_unique_id
```
