Perform element-wise math, use broadcasting, and apply universal functions.

```python
# Import the NumPy library and alias it as 'np'
import numpy as np
```

## 1. Element-wise Operations

```python
# Create two arrays for demonstration
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])

# Add the two arrays element by element
add_result = arr1 + arr2
print(f"Addition: {add_result}")

# Multiply the two arrays element by element
mult_result = arr1 * arr2
print(f"Multiplication: {mult_result}")

# Square each element in arr1 (raise to the power of 2)
pow_result = arr1 ** 2
print(f"Power (Squared): {pow_result}")
```

## 2. Broadcasting

```python
# Create a 2x3 matrix of ones
matrix = np.ones((2, 3))
print(f"Original Matrix:\n{matrix}")

# Add a scalar value (5) to the entire matrix. NumPy 'broadcasts' the 5 to match the matrix shape.
broadcast_result = matrix + 5
print(f"Matrix + 5:\n{broadcast_result}")
```

## 3. Universal Functions (ufuncs) & Aggregation

```python
# Create an array of values [1, 4, 9, 16]
arr = np.array([1, 4, 9, 16])

# Calculate the square root of each element using np.sqrt()
print(f"Square Root: {np.sqrt(arr)}")

# Calculate the sum of all elements in the array using np.sum()
print(f"Sum: {np.sum(arr)}")

# Find the maximum value in the array using np.max()
print(f"Max Value: {np.max(arr)}")

# Find the mean (average) of the array using np.mean()
print(f"Mean: {np.mean(arr)}")
```
