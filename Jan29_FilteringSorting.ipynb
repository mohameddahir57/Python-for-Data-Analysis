In this notebook, we'll learn how to slice and dice our data to find exactly what we need.

```python
import pandas as pd
```

### 1. Load Data
Load the `practice_data.csv`.

```python
df = pd.read_csv('practice_data.csv')
df
```

### 2. Basic Filtering
Filter the dataframe to show only products with a Price > 50.

```python
expensive = df[df['Price'] > 50]
expensive
```

### 3. Multiple Conditions
Find products that are NOT 'Laptop' AND have a price greater than 30.

```python
subset = df[(df['Product'] != 'Laptop') & (df['Price'] > 30)]
subset
```

### 4. Sorting
Sort the dataframe by Price in descending order.

```python
sorted_df = df.sort_values(by='Price', ascending=False)
sorted_df
```
