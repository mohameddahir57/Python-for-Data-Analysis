In this notebook, we will practice creating and using the two main Pandas structures: **Series** and **DataFrames**.

```python
import pandas as pd
import numpy as np
```

### 1. Creating a Series
Create a Series named `fruits` with the values: 'Apple', 'Banana', 'Cherry' and index them as 'a', 'b', 'c'.

```python
data = ['Apple', 'Banana', 'Cherry']
labels = ['a', 'b', 'c']

fruits = pd.Series(data, index=labels)
print(fruits)
```

### 2. Creating a DataFrame
Create a DataFrame named `df_students` from the following dictionary.

```python
student_data = {
    'Student': ['Ali', 'Sara', 'John'],
    'Score': [85, 92, 78],
    'Subject': ['Math', 'Science', 'History']
}

df_students = pd.DataFrame(student_data)
df_students
```

### 3. Basic Inspection
Use `.head()`, `.shape`, and `.columns` to inspect `df_students`.

```python
print("Head:\n", df_students.head())
print("\nShape:", df_students.shape)
print("\nColumns:", df_students.columns)
```
