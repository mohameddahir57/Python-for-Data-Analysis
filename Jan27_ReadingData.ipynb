In this notebook, we'll explore how to load data from standard file formats.

```python
import pandas as pd
```

### 1. Creating a dummy CSV for practice
Run the cell below to create a small CSV file named `practice_data.csv`.

```python
data = """ID,Product,Price
1,Laptop,1000
2,Mouse,25
3,Keyboard,50
4,Monitor,200
"""
with open('practice_data.csv', 'w') as f:
    f.write(data)
print("File 'practice_data.csv' created!")
```

### 2. Loading the CSV
Read the `practice_data.csv` file into a DataFrame named `df`.

```python
df = pd.read_csv('practice_data.csv')
df
```

### 3. Basic Inspection After Loading
Check the data types of the columns and the summary statistics.

```python
print("Data Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe())
```
