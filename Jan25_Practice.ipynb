**Scenario:** You have a small dataset representing the daily sales (in USD) of a store for one week. You also have a second dataset for the following week. Your goal is to analyze performance.

```python
# Import NumPy
import numpy as np
```

### Problem 1: Create Data

```python
# Create an array 'week1_sales' with values: [200, 220, 250, 300, 280, 400, 450]
week1_sales = np.array([200, 220, 250, 300, 280, 400, 450])
print(f"Week 1 Sales: {week1_sales}")

# Create an array 'week2_sales' with values: [210, 230, 260, 310, 290, 420, 480]
week2_sales = np.array([210, 230, 260, 310, 290, 420, 480])
print(f"Week 2 Sales: {week2_sales}")
```

### Problem 2: Basic Stats

```python
# Calculate total sales for Week 1 using np.sum()
total_week1 = np.sum(week1_sales)
print(f"Total Week 1 Sales: ${total_week1}")

# Calculate average daily sales for Week 1 using np.mean()
avg_week1 = np.mean(week1_sales)
print(f"Average Week 1 Sales: ${avg_week1:.2f}")

# Find the highest daily sales in Week 1 using np.max()
max_week1 = np.max(week1_sales)
print(f"Highest Single Day Sales: ${max_week1}")
```

### Problem 3: Comparison

```python
# Calculate the difference in sales between Week 2 and Week 1 (element-wise subtraction)
growth = week2_sales - week1_sales
print(f"Daily Growth (Week 2 - Week 1): {growth}")

# Calculate the total growth amount
total_growth = np.sum(growth)
print(f"Total Growth: ${total_growth}")
```

### Problem 4: Filtering

```python
# Find all days in Week 1 where sales were greater than $300
# Create a boolean mask
high_sales_mask = week1_sales > 300

# Use the mask to get the actual values
high_sales_days = week1_sales[high_sales_mask]
print(f"Days with sales > $300 in Week 1: {high_sales_days}")
```
