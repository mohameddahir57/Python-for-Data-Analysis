In this notebook, we will practice identifying and handling missing values using a sample dataset.

```python
import pandas as pd
import numpy as np

# Create a sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age': [25, np.nan, 30, 35, np.nan, 40],
    'City': ['New York', 'London', np.nan, 'Paris', 'New York', 'London'],
    'Salary': [50000, 60000, 70000, np.nan, 55000, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
df
```

### 1. Check for missing values

```python
print("Missing values per column:")
print(df.isna().sum())
```

### 2. Filling missing values with a constant

```python
# Fill City with 'Unknown'
df['City'] = df['City'].fillna('Unknown')
df
```

### 3. Imputation (Filling with Mean/Median)

```python
# Fill Age with Median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Salary with Mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print("DataFrame after filling:")
df
```

### 4. Dropping remaining missing values (if any)

```python
# In this case, we've filled everything, but here is how you'd drop
df_cleaned = df.dropna()
df_cleaned
```
